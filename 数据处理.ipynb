### ml-1m 序列数据

```python
import time
import datetime
```

```python
def getTime(timeVal):
  t = datetime.datetime.strptime(timeVal, '%Y-%m-%d')
  d = t.timetuple()
  return int(time.mktime(d))

def getDate(timestamp):
  timeArray = time.localtime(timestamp)
  timestr = time.strftime("%Y-%m-%d %H:%M:%S", timeArray)
  date = datetime.datetime.strptime(timestr, "%Y-%m-%d %H:%M:%S")
  return date
```

### 处理评分数据

```python
rating_data = pd.read_table('ratings.dat', sep='::', header=None, names=['userId', 'movieId', 'rating', 'timestamp'])
```

```python
rating_data[:4]
```

```python
rating_data['movie'] = rating_data.movieId.map(movieId_dict)
rating_data['user'] = rating_data.userId.map(userId_dict)
```

```python
rating_data.head()
```

```python
max_date = getDate(max(rating_data.timestamp))
min_date = getDate(min(rating_data.timestamp))
```

```python
rating_data['month'] = pd.cut(rating_data.timestamp, cut_31Data(), labels=list(range(31))).astype(np.int64)
```

```python
rating_data['day'] = rating_data.timestamp.map(lambda x : (getDate(x)-min_date).days)
```

```python
def handle_data(udata):
    
    len_udata = len(udata)
    window = int(len_udata*0.1)
    
    if window < 10:
        window =10
    elif window > 50:
        window = 50
    
    test_i = udata[-1]
    test_input = udata[-1-window:-1]
    
    train_inputs = []
    train_is = []
    for index, i in enumerate(udata[window:-1]):
        train_is.append(i)
        train_inputs.append(udata[index:index+window])
    
    return train_inputs, train_is, test_i, test_input
    
```
